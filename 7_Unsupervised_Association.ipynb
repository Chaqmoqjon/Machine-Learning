```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
dataset = pd.read_csv("/content/Market_Basket_Optimisation.csv", header=None)
dataset.head()
```

```python
dataset.shape
```

```python
#hamma transaksiyalar yuklandi
transactions = []
for i in range(0, len(dataset)):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

##Apiori

```python
#kerakli mahsulotlarni ajratish qoidalari
rules = apriori(transactions = transactions, min_support = 0.003, min_cinfidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(rules)
results
```

```python
#df ga aylantirildi
def inspect(results):
    lhs         =[tuple(result[2][0][0])[0] for result in results]
    rhs         =[tuple(result[2][0][1])[0] for result in results]
    supports    =[result[1] for result in results]
    confidences =[result[2][0][2] for result in results]
    lifts        =[result[2][0][3] for result in results]
    return list (zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ["Left hand side", "Right hand side", "Support", "Confidence", "Lift"])
resultsinDataFrame
```

```python
#tekshirildi
resultsinDataFrame.nlargest(n = 10, columns = "Lift")
```

##Eclat

```python
# Mahsulotlarning tez-tez birga sotilish to'plamlarini hisoblash (ECLAT usuli)
from collections import defaultdict
```

```python
# Har bir mahsulot kombinatsiyasining chastotasi
itemsets = defaultdict(int)
for transaction in transactions:
    for item in set(transaction):  # Qaytarma elementlar hisobga olinmasligi uchun 'set' ishlatildi
        itemsets[item] += 1
```

```python
# Minimal support darajasi bo'yicha filtr qilish
min_support = 0.003
min_count = min_support * len(transactions)
```

```python
# ECLAT natijalari
eclat_results = {item: count for item, count in itemsets.items() if count >= min_count}

# DataFrame yaratish
eclat_df = pd.DataFrame(list(eclat_results.items()), columns=["Item", "Support"])
eclat_df["Support"] = eclat_df["Support"] / len(transactions)  # Qo'llab-quvvatlashni foizga aylantirish
```

```python
# Support bo'yicha eng katta 10 ta mahsulotni ko'rish
eclat_df.nlargest(n=10, columns="Support")
```

##F-P Growth Algorithm

```python
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```python
# Mahsulotlar ro'yxatini list formatiga o'tkazish
all_items = list(set(item for transaction in transactions for item in transaction if item != 'nan'))

# Tranzaksiyalarni binar formatda kodlash uchun DataFrame yaratish
encoded_data = pd.DataFrame(0, index=np.arange(len(transactions)), columns=all_items)


# Tranzaksiyalarni binar formatda kodlash
for i, transaction in enumerate(transactions):
    for item in transaction:
        if item != 'nan':
            encoded_data.loc[i, item] = 1
```

```python
# F-P Growth algoritmini qo'llash
frequent_itemsets = fpgrowth(encoded_data, min_support=0.003, use_colnames=True)

# Assotsiatsiya qoidalarini yaratish
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=3)

# Natijalarni ko'rsatish
rules_df = rules[["antecedents", "consequents", "support", "confidence", "lift"]]

# Support bo'yicha eng yuqori qoidalarni ko'rsatish
rules_df.nlargest(n=10, columns="lift")
```
